```python
import matplotlib.pyplot as plt
```

```python
# Basic plots
# Line plot
X = [1,2,3,4,5]
Y = [2,4,6,8,10]

X1 = [10,20,30,40,50]
Y1 = [1,2,3,4,5]

#plt.plot(X,Y)
#plt.plot(X,Y,color='red')
#plt.plot(X,Y,color='red',marker='o')
#plt.plot(X,Y,color='red',marker='o',linestyle='--')
#plt.plot(X1,Y1)
plt.plot(X,Y,X1,Y1,color='red',marker='o',linestyle='--')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

```

```python
#scatter plot

X = [1,2,3,4,5]
Y = [2,4,6,8,10]

plt.scatter(X,Y)
plt.grid(True)
plt.title("scatter plot")
```

```python
# Bar plot
X = ['A','B','C','D']
Y = [10,20,12,30]

plt.bar(X,Y)
```

```python
# Histogram
data = [22,45,62,23,45,62,74,89,62,45]

#plt.hist(data)
plt.hist(data,bins=5)
```

```python
# pie chart
labels = ['A','B','C','D']
values = [50,20,15,15]

#plt.pie(values)
plt.pie(values,labels = labels,autopct='%0.2f%%')
```

```python
# subplots

plt.subplot(1,2,1)
plt.plot([1,2,3],[4,5,6])

plt.subplot(1,2,2)
plt.pie(values,labels = labels,autopct='%0.2f%%')

```
