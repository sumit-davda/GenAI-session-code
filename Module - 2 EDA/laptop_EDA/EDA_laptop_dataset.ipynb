```python
df['ScreenResolution'].value_counts()
```

```python
def isTouchScreen(name):
    if 'Touchscreen' in name:
        return 1
    else:
        return 0
```

```python
#df['TouchScreen1'] = df['ScreenResolution'].apply(isTouchScreen)
# Ternary operator : true_stat if else false_stat
df['TouchScreen'] = df['ScreenResolution'].apply(lambda name : 1 if  'Touchscreen' in name else 0)
```

```python
df['TouchScreen'].value_counts().plot(kind='bar')
```

```python
sns.barplot(x='TouchScreen',y='Price',data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
df['IPS'] = df['ScreenResolution'].apply(lambda name : 1 if  'IPS' in name else 0)
df['IPS'].value_counts().plot(kind='bar')
```

```python
sns.barplot(x='IPS',y='Price',data=df)
plt.xticks(rotation='vertical')
plt.show()
```
